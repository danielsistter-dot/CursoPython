```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
engine =create_engine("sqlite:///ecommerce.db")
```

```python
df=pd.read_csv("dados/produtos.csv")
df
```

```python
df.head()
```

```python
try:
    df.to_sql('produtos', con=engine,if_exists='replace',index=False)
    print("Dados importados com sucesso")
except Exception as e:
    print(f"Ocorreu uma falha na importação {e}")
```

```python
consulta_1 ="SELECT Categoria, SUM(quantidade) AS Total_Qtd_Vendida FROM produtos GROUP BY Categoria"
Resultado_1 = pd.read_sql_query(consulta_1,engine)
Resultado_1
```

```python
consulta_2="SELECT Produto, SUM(Quantidade) AS Total_Vendido FROM produtos GROUP BY produto ORDER BY Total_Vendido DESC LIMIT 1"
Resultado_2 = pd.read_sql_query(consulta_2,engine)
Resultado_2
```

```python
consulta_3="SELECT Produto, SUM(Quantidade) AS Quantidade_Vendida from produtos GROUP BY Produto ORDER BY Quantidade_Vendida DESC"
Resultado_3=pd.read_sql_query(consulta_3, engine)
Resultado_3
```

```python
consulta_4="SELECT Produto, AVG (Preço) AS Media_de_Preco FROM produtos GROUP BY  Produto"
Resultado_4= pd.read_sql_query(consulta_4, engine)
Resultado_4
```
