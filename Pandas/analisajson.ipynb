```python
import pandas as pd
```

```python
df=pd.read_json("dados/receitas.json", encoding='utf-8')
df
```

```python
receita_expandidas = []
for receita in df['Receitas']:
    total_ingredientes = len(receita['Ingredientes'])
    receita_info = {
        'Titulo': receita['Titulo'],
        'Tempo de Preparo':receita["Tempo de Preparo"],
        'Porções':receita['Porcoes'],
        'Total de Ingredientes':total_ingredientes
    }
    receita_expandidas.append(receita_info)
```

```python
receita_expandidas
```

```python
df_expandido = pd.DataFrame(receita_expandidas)
df_expandido
```
