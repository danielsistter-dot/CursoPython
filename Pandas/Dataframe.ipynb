```python
dados_notas = {
    'Nome':['Alice','Bob','Charli','David','Eva'],
    'Matematica': [85,90,78,88,92],
    'Fisica': [80,75,88,85,95],
    'Quimica':[78,82,85,80,88]
}
```

```python
df_notas=pd.DataFrame(dados_notas)
df_notas
```

```python
df_notas['Media']= df_notas[['Matematica','Fisica','Quimica']].mean(axis=1)
df_notas
```

```python
aluno_aprovado =df_notas[df_notas['Media']>=82.5]
aluno_aprovado
```
