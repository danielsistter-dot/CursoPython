```python
nomes = np.array(["Alice","Bob","Charli"])
notamatematica = np.array ([85,90,78])
notafisica = np.array([88,76,95])
```

```python
dadosestudantes =np.vstack((nomes, notamatematica, notafisica))
dadosestudantes
```

```python
dadosestudantesh= np.column_stack(( notamatematica,notafisica))
dadosestudantesh
```

```python
medianotas =  np.mean(dadosestudantesh, axis=1)
final = np.column_stack((nomes,dadosestudantesh,medianotas))
final
```
